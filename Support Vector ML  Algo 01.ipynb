# Q4. What is the role of support vectors in SVM Explain with example

 Support vectors are the data points that lie closest to the hyperplane and influence the position and orientation of the hyperplane. Support Vector play a major role in Svm , which are illuserated one by one in above section:
 
 **Defining the Decision Boundary :**  The hyperplane is determined by the support vectors, and it is positioned to have the maximum margin while still correctly classifying all training data.
 
 **Determining the Margin :** As the decision boundray are determined by the support vactor and the distance between the decision baoundry is called margin, so the Support vector has a big role to defining the Margin between planes.
 
 **Influence on the Weight Vector:** The support vectors contribute to determining the orientation of the hyperplane and are essentially the "building blocks" of the decision boundary.
 
 **Robustness and Generalization:**   They represent the instances that are most relevant to the decision boundary and, as a result, contribute significantly to the model's robustness and generalization to unseen data.

# Q6. SVM Implementation through Iris dataset.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading the dataset
df=sns.load_dataset('iris')
```

```python
#defining the variables
x= df.iloc[:,: -1]
y= df.iloc[::,-1]
```

```python
# spliting the data into traing and test data

from sklearn.model_selection import train_test_split
x_train, x_test , y_train,y_test = train_test_split (x,y , test_size= 0.33 , random_state=42)
```

```python
#importing the libarary linear Classifier
from sklearn.svm import LinearSVC

#creating the object of liner svc
svc= LinearSVC()

#fitting the data to object for training
svc.fit(x_train , y_train)

#predictsion for y
y_pred = svc.predict(x_test)
```

```python
# calculating the accuracy of the model.

from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

# Trying on different values of the regularisation parameter C

```python
# The default value of perameter C is 1 .

# we are trying with 3

svc= LinearSVC(C= 3)
svc.fit(x_train,y_train)
y_pred= svc.predict(x_test)
print(accuracy_score(y_pred , y_test))

print('accuracy remains same')
```
